#Distribuci√≥n de Probabilidad

##Importar bibliotecas de funciones importantes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##Importar el dataset - copiado al folder por facilidad

```python
df = pd.read_csv('wine.data')
columns = ['class', 'Alcohol', 'Malic Acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
df.columns = columns
df.head()
```

```python
import seaborn as sns
sns.set_theme()
```

```python
class_1 = df[0:58]
class_2 = df[58:129]
class_3 = df[129:178]
```

##Caracter√≠stica: Alcohol
##Distribuci√≥n aproximada: Uniforme

```python
sns.histplot(data = df, x = 'Alcohol', hue = "class", bins = 30, multiple="stack")
```

```python
sns.distplot(df['Alcohol'], kde = True, color ='red', bins = 30)
```

##Gr√°ficas para entender mejor c√≥mo funcionan los histogramas

```python
sns.histplot(data = class_1, x = 'Alcohol', hue = "class", bins = 30)
```

```python
sns.histplot(data = class_2, x = 'Alcohol', hue = "class", bins = 30)
```

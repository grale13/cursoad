```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sweetviz as sv
```

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
report = sv.analyze(df_train)
report.show_html()
```

## Comparison across subsets (Train/Test)

```python
feature_config = sv.FeatureConfig(skip="PassengerId", force_text=['Age'])
report = sv.compare([df_train, "Training Data"], [df_test, "TestData"], "Survived", feature_config)
```

```python
report.show_html("Titanic.html", open_browser = True)
```

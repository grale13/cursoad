```python
def factorial(n):
  if n == 1: return 1
  else: return n * factorial(n - 1)

factorial(4)  
```

```python
def permutation_norep(n,r):
  return factorial(n) / factorial(n - r)

permutation_norep(36,8)
```

##With repetition

```python
def permutation_withrep(n,r):
  return n ** r

permutation_withrep(36,8)
```

#Combinations

## Without repetition

```python
def combination_norep(n,r):
  return factorial(n) / (factorial(r) * factorial(n - r))

combination_norep(10,5)
```

##With repetition

```python
def combination_withrep(n,r):
  return factorial(n + r - 1) / (factorial(r) * factorial(n - 1))

combination_withrep(10,4)
```

#Flip coins

```python
import random

def flip_coin(n_simulations = 10):
  # aguila 0
  # sol 1

  count_aguila = 0
  count_sol = 0

  for i in range(n_simulations):
    result = random.randint(0,1)

    #if aguila
    if result == 0:
      count_aguila += 1

    #else sol
    else:
      count_sol += 1

  return [count_aguila, count_sol]



```

```python
count_aguila, count_sol = flip_coin(10)
print ("#aguila: ", count_aguila, "#sol: ", count_sol)
```

#Dice simulation

```python
import random

def roll_dice(n_simulations = 10):
  # aguila 0
  # sol 1

  count = 0
  
  for i in range(n_simulations):
    dice1 = random.randint(1,6)
    dice2 = random.randint(1,6)

    score = dice1 + dice2

    if score == 7:
      count += 1

  return count/n_simulations

```

```python
print("resultado: ", roll_dice(10))
```

```python
n_simulations = [1, 2, 4, 8, 16, 32, 64, 128, 256]
probs = []

for n in n_simulations:
  prob = roll_dice(n)
  probs.append(prob)
  print(n_simulations, " - ", prob)

```

```python
import matplotlib.pyplot as plt

truth = 6/36

plt.figure(figsize=(18, 3))

# plot the theoretical value
plt.plot(n_simulations, [truth] * len(n_simulations), c = "r", label = "Theoretical Probability")

# plot simulation (experiments)
plt.plot(n_simulations, probs, c = "b", label = "Simulation Probability")

plt.xlabel("N simulations")
plt.ylabel("P(X = 7)")
plt.title("Roll the dice simulation - P(X = 7)")
plt.legend()
plt.show()
```

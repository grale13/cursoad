#Uniform distribution

```python
from scipy.stats import uniform

n = 100000
a = 10 # start
b = 20 # scale: a + b

# create fake data
data_uniform = uniform.rvs(size = n, loc = a, scale = b)
```

```python
from seaborn.distributions import distplot
ax = sns.distplot(data_uniform, bins = 20, color = 'b', kde = False)
ax.set(xlabel='Uniform distribution' , ylabel='Frecuency')
```

```python
ax = sns.distplot(data_uniform, bins = 20, color = "b", kde = True)
ax.set(xlabel="Uniform Distribution", ylabel="Probability")
```

```python
from scipy.stats import norm

n = 100000
a = 0
b = 1

# create fake data
# loc: mean, scale: stdv
data_norm = norm.rvs(size = n, loc = a, scale = b)

ax = sns.distplot(data_norm, bins = 20, color = "b", kde = False)
ax.set(xlabel="Normal Distribution", ylabel="Freq.")
```

```python
from scipy.stats import expon

n = 100000
a = 0
b = 1

# create fake data
# loc: mean, scale: 1/lambda
data_expon = expon.rvs(size = n, loc = a, scale = b)

ax = sns.distplot(data_expon, bins = 20, color = "b", kde = True)
ax.set(xlabel="Exponential Distribution", ylabel="Prob.")
```

```python
from scipy.stats import poisson

n = 100000
mu = 3
# create fake data
data_poisson = poisson.rvs(size = n, mu = mu)

ax = sns.distplot(data_poisson, bins = 30, color = "b", kde = False)
ax.set(xlabel="Poisson Distribution", ylabel="Freq.")
```

```python
from scipy.stats import bernoulli

n = 100000
p = 0.6

# create fake data
data_bernoulli = bernoulli.rvs(size = n, p = p)

ax = sns.distplot(data_bernoulli, bins = 10, color = "b", kde = False)
ax.set(xlabel="Bernoulli Distribution", ylabel="Freq.")
```
